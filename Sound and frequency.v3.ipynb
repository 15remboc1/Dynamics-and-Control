```python
from IPython.display import Audio
```

```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

We can play a note in the notebook

```python
fs = 44100 # sampling frequency
T = .5    # seconds
A = 440
twopi = 2*numpy.pi

t = numpy.linspace(0, T, int(T*fs), endpoint=False) # time variable
def note(frequency):
    return numpy.sin(twopi*frequency*t)                # pure sine wave at 440 Hz

# load a NumPy array
Audio(note(A), rate=fs)
```

Chromatic scale

```python
scalet = numpy.linspace(0, T*13, int(T*fs*13), endpoint=False)
scale = numpy.concatenate([note(A*2**(i/12.)) for i in range(13)])    
```

```python
Audio(scale, rate=fs)
```

Of course, this is quite a large number of points, you can't really see what's going on

```python
plt.plot(scalet, scale)
```

You can see the sinusoids better if you zoom in

```python
Nzoom = 1000
plt.plot(scalet[:Nzoom], scale[:Nzoom])
```

Let's listen to the effect of running this through a first order filter. We'll use the convolution.

```python
omega = 440/twopi
tau = 1/omega
first_order_impulse = numpy.exp(-t/tau)
```

```python
plt.plot(first_order_impulse)
```

```python
filtered = numpy.convolve(scale, first_order_impulse)
```

```python
Audio(filtered[:len(scalet)], rate=fs)
```

Can you hear how the sound gets softer as the frequency goes up? This is the effect of the filter.
