Padé approximation
==========

Padé approximation is an extension of the concept of Taylor approximation with rational functions rather than polynomials. The basic premise is that the approximation is made by matching derivatives.

```python
import sympy
sympy.init_printing()
%matplotlib inline
```

```python
import numpy
```

```python
a, b, c, s, D = sympy.symbols('a, b, c, s, D')
```

We will approximate a Laplace dead time

```python
G = sympy.exp(-2*s)
G
```

by a 1/1 Padé approximation. This means first order above the line and first order below. In order to force uniqueness of the solution, we force the constant term in the denominator to be unity.

```python
ga = (a + b*s)/(1 + c*s)
ga
```

We can solve for the parameters by setting the function value, first derivative and second derivative of the original function and the approximation to be equal at $s=0$.

```python
soln = sympy.solve([(G.diff(s, i) - ga.diff(s, i)).subs({s: 0}) for i in range(3)], 
                   [a, b, c], dict=True)
soln
```

```python
G_pade = ga.subs(soln[0])
G_pade
```

So how much do the approximations resemple the original function?

```python
def cplot(G):
    f = sympy.lambdify(s, G, ['mpmath', 'numpy'])
    sympy.mpmath.cplot(f, [-5, 5], [-5, 5])
```

The original function

```python
cplot(G)
```

Pade approximation

```python
cplot(G_pade)
```

Taylor approximation with same number of coefficients

```python
G_taylor = G.series(s, 0, 4).removeO()
G_taylor
```

```python
cplot(G_taylor)
```
